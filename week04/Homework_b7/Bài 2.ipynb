# Chuẩn bị môi trường


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.metrics import precision_score, recall_score, f1_score

pd.set_option("display.max_columns", None)
sns.set_theme(style="whitegrid")
np.random.seed(42)          # cố định ngẫu nhiên -> kết quả tái lập được
print("Sẵn sàng.") 
```

# Đọc dữ liệu

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

DATA_PATH = Path("Dry_Bean_Dataset") / "Dry_Bean_Dataset.xlsx"

OUTPUT_DIR = DATA_PATH.parent
TRAIN_OUTPUT_PATH = OUTPUT_DIR / "dry_bean_train.csv"
TEST_OUTPUT_PATH = OUTPUT_DIR / "dry_bean_test.csv"


train_dts = pd.read_csv(TRAIN_OUTPUT_PATH)
x_train = train_dts.drop(columns="class")
y_train = train_dts["class"]

test_dts = pd.read_csv(TEST_OUTPUT_PATH)
x_test = test_dts.drop(columns="class")
y_test = test_dts["class"]
```

# LOGICTIS REGRESSION

```python

model = LogisticRegression()
model.fit(x_train,y_train)
y_pred_lr = model.predict(x_test)
y_pred_lr.reshape(len(y_pred_lr),1)

print(y_pred_lr.shape)
print(y_test.shape)
precision = precision_score(y_test, y_pred_lr, average="macro")
recall = recall_score(y_test, y_pred_lr, average="macro")
f1 = f1_score(y_test, y_pred_lr, average="macro")

print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score, recall_score, f1_score

param_grid = {
    "n_neighbors": [10,20,40,80,100],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

grid_search = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=10,
    scoring="accuracy"
)

grid_search.fit(x_train, y_train)

best_knn = grid_search.best_estimator_

y_pred = best_knn.predict(x_test)

print(grid_search.best_params_)

precision = precision_score(y_test, y_pred, average="macro")
recall = recall_score(y_test, y_pred, average="macro")
f1 = f1_score(y_test, y_pred, average="macro")

print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1-score  : {f1:.4f}")
```

# Do là KNN có precision,recall,f1-score đều cao hơn, nên là với loại bài này nên dùng KNN :)
